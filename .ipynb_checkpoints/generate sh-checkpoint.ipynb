```python
import os
os.remove("template.sh")

seed_list = [2]
model_list = ["tinyhar"]#["deepconvlstm_attn","deepconvlstm","attend"] # deepconvlstm_attn  attend
data_list  = ["hapt", "dg" ,"wisdm", "dsads" ,"oppo" ,"pamap2","rw"]#["hapt", "dg" ,"wisdm", "dsads" ,"oppo" ,"pamap2"] #"hapt" "dg" "wisdm" "dsads" "oppo" "pamap2"
defference_list = [True, False]
filtering_list = [ True,False]

filter_scaling_factor = 1.0
mixup_argmax_list = [True]
mixup_probability_list = [0.5,1]
random_augmentation_prob_list = [0.5,1]

with open("template.sh",'a') as f:
    for seed in seed_list:
        for model in model_list:
            for data in data_list:
                for filtering in filtering_list:
                    for mixup_argmax in mixup_argmax_list:
                        for mixup_probability in mixup_probability_list:
                            for random_augmentation_prob in random_augmentation_prob_list:

                                if filtering:
 
                                    f.write(
                                        "sbatch main.py --data-name {} --shuffle --model-type {} --seed {} --difference --filtering --mixup-argmax --root-path ../../../datasets --filter-scaling-factor {} --mixup-probability {} --random-augmentation-prob {}".format(
                                                                   data,                     model,    seed    ,                                                           filter_scaling_factor ,      mixup_probability   ,random_augmentation_prob)
                                    )
                                    f.write('\n')  
                                else:

                                    f.write(
                                        "sbatch main.py --data-name {} --shuffle --model-type {} --seed {}  --mixup-argmax --root-path ../../../datasets --filter-scaling-factor {} --mixup-probability {} --random-augmentation-prob {}".format(
                                                                   data,                     model,    seed      ,                                       filter_scaling_factor    ,         mixup_probability    ,      random_augmentation_prob)
                                    )
                                    f.write('\n')    
```
